```python

def get_source(profile_url):
    driver.get(profile_url)
        
    scroll_down()
        
    src = driver.page_source
    soup = BeautifulSoup(src, "html.parser")
    return soup

# Scrap Url
def scrap_url():
    profile_url = driver.current_url
    parts = profile_url.split("/")
    url = '/'.join(parts[:5])
    return url

#  Id Generator
def id_generator(name):
    global count
    total_digits = 7
    current_datetime = datetime.now()
    time = current_datetime.strftime("%H%M%S")
    segments = name.split(" ")
    username = segments[0]
    username = username[0:3]
    
    count = count + 1
    number = "0" * (total_digits - len(str(count))) + str(count)
    id = time + username.upper() + number
    return id

# Scrolling Down   
def scroll_down():
    start = time.time()

    initialScroll = 0
    finalScroll = 1000

    while True:
        driver.execute_script(f"window.scrollTo({initialScroll},{finalScroll})")
        initialScroll = finalScroll
        finalScroll += 1000
        
        time.sleep(3)
        
        end = time.time()
        if round(end - start) > 20:
            break


# Scrapping Name
def scrap_name(soup):
    
    name_div = soup.find('h1', {'class' : 'text-heading-xlarge'})
    name = name_div.text
        
    return name
    
# Scrapping Summary
def scrap_summary(soup):
    summary_div = soup.find('div', {'class' : 'text-body-medium break-words'})
    if summary_div:
        summary = summary_div.get_text().strip()
    else:
        summary = ''
    return summary

# Scrapping Location
def scrap_loc(soup):
    loc_select = f'div > div.scaffold-layout.scaffold-layout--breakpoint-none.scaffold-layout--main-aside.scaffold-layout--single-column.scaffold-layout--reflow.pv-profile > div > div > main > section.artdeco-card.pv-top-card > div.ph5.pb5 > div.mt2.relative > div.GAQcIWUkkZnxtXekLnyavAPJKwOWMTnmp.mt2 > span.text-body-small.inline.t-black--light.break-words'
    loc_div = soup.select_one(loc_select)
    if loc_div:
        loc = loc_div.text
        loc = loc.replace('Contact Info', '')
    else:
        loc_div = soup.select_one(f'div > div.scaffold-layout.scaffold-layout--breakpoint-none.scaffold-layout--main-aside.scaffold-layout--single-column.scaffold-layout--reflow.pv-profile > div > div > main > section.artdeco-card.pv-top-card > div.ph5.pb5 > div.mt2.relative > div.GAQcIWUkkZnxtXekLnyavAPJKwOWMTnmp.mt2 > span.text-body-small.inline.t-black--light.break-words')
        if loc_div:
            loc = loc_div.text
        else:
            loc = ''
    return loc

# Scrapping About
def scrap_about(soup):
    about_div = soup.find('div', {'class': 'display-flex ph5 pv3'})
    if about_div:
        about = about_div.get_text(strip=True)

        # Removing the "See more" link text
        about = about.replace('…see more', '')
    else:
        about = ''
    return about        


# Adding Profile data to CSV
def add_profileData_to_csv(id, name, summary, loc, url, about, url2):
    field_names = ['ID', 'Name', 'Summary', 'Location', 'URL', 'About', 'URL 2']

    data = {
        'ID': id,
        'Name': name,
        'Summary' : summary,
        'Location': loc,
        'URL': url,
        'About': about,
        'URL 2' : url2
    }

    try:
        with open('profile_data.csv', mode='a', newline='', encoding='utf-8') as file:  # Specify encoding as UTF-8
            writer = csv.DictWriter(file, fieldnames=field_names)

            # Check if the file is empty and write headers if needed
            file.seek(0, 2)
            file_empty = file.tell() == 0

            if file_empty:
                writer.writeheader()

            writer.writerow(data)
            print("Data added to CSV successfully.")
    except IOError:
        print("Error: Could not write to CSV file.")

def exists_in_csv(experience):
    with open('experience_data.csv', mode='r', encoding='utf-8') as file:
        reader = csv.reader(file)

        for row in reader:
            if experience in row:
                return True

    return False



# Getting URLs
def main():
    with open('cleanedUrls.csv', mode='r', encoding='utf-8') as file:
        csv_reader = csv.DictReader(file)
        
        for index, row in enumerate(csv_reader):
            if index < 684:
                continue
            
            profile_url = row['URLs']

            # Calling Scrapping Functions
            #time.sleep(30)
            driver.get(profile_url)
            soup = get_source(profile_url)
            name = scrap_name(soup)
            id = id_generator(name)
            #scrap_experience(id)
            url = scrap_url()
            time.sleep(5)
            loc = scrap_loc(soup)
            about = scrap_about(soup)
            time.sleep(8)
            summary = scrap_summary(soup)
                
            add_profileData_to_csv(id, name, summary, loc, url, about, profile_url)

        
main()
```
